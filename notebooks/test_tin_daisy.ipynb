```python
from collections import Counter
counts = Counter(bammap['case'])
case_to_dtypes = {}
for i, row in bammap.iterrows():
    d = {}
    for c in ['case', 'disease', 'experimental_strategy', 'sample_type', 'data_path', 'filesize']:
        d[c] = row[c]
    d['# sample_name'] = i
    
    if row['case'] not in case_to_dtypes:
        case_to_dtypes[row['case']] = {}
        
    case_to_dtypes[row['case']][i] = d
```

```python
for case, count in counts.most_common():
    print(case, count,
          len([x for x in case_to_dtypes[case].keys() if 'WXS' in x]),
          len([x for x in case_to_dtypes[case].keys() if 'WGS' in x]),
          len([x for x in case_to_dtypes[case].keys() if 'RNA' in x]))
```
